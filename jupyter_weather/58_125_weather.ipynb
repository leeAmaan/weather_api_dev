```python
import pandas as pd
import datetime
```

```python
RN1_data = pd.read_csv("C:/Users/happytoseeyou/Desktop/58_125/가산동_강수_202201_202210.csv", encoding="utf-8-sig")
PTY_data = pd.read_csv("C:/Users/happytoseeyou/Desktop/58_125/가산동_강수형태_202201_202210.csv", encoding="utf-8-sig")
T1H_data = pd.read_csv("C:/Users/happytoseeyou/Desktop/58_125/가산동_기온_202201_202210.csv", encoding="utf-8-sig")
REH_data = pd.read_csv("C:/Users/happytoseeyou/Desktop/58_125/가산동_습도_202201_202210.csv", encoding="utf-8-sig")
WSD_data = pd.read_csv("C:/Users/happytoseeyou/Desktop/58_125/가산동_풍속_202201_202210.csv", encoding="utf-8-sig")
#SKY_data = pd.read_csv("C:/Users/happytoseeyou/Desktop/20221128100305/청운효자동_하늘상태_202201_202210.csv", encoding="utf-8-sig")
```

```python
RN1_data.columns = ["day", "baseTime", 'RN1']
PTY_data.columns = ["day", "baseTime", 'PTY']
T1H_data.columns = ["day", "baseTime", 'T1H']
REH_data.columns = ["day", "baseTime", 'REH']
WSD_data.columns = ["day", "baseTime", 'WSD']
#SKY_data.columns = ["day", "baseTime", 'SKY']
```

```python
RN1_data = RN1_data[~RN1_data['day'].str.contains('Start')]
PTY_data = PTY_data[~PTY_data['day'].str.contains('Start')]
T1H_data = T1H_data[~T1H_data['day'].str.contains('Start')]
REH_data = REH_data[~REH_data['day'].str.contains('Start')]
WSD_data = WSD_data[~WSD_data['day'].str.contains('Start')]
#SKY_data = SKY_data[~SKY_data['day'].str.contains('Start')]
```

```python
RN1_data['baseDate'] = pd.date_range(start='1/1/2022', periods=len(RN1_data), freq='H')
PTY_data['baseDate'] = pd.date_range(start='1/1/2022', periods=len(PTY_data), freq='H')
T1H_data['baseDate'] = pd.date_range(start='1/1/2022', periods=len(T1H_data), freq='H')
REH_data['baseDate'] = pd.date_range(start='1/1/2022', periods=len(REH_data), freq='H')
WSD_data['baseDate'] = pd.date_range(start='1/1/2022', periods=len(WSD_data), freq='H')
#SKY_data['baseDate'] = pd.date_range(start='1/1/2022', periods=len(SKY_data), freq='H')
```

```python
RN1_data['baseTime'] = RN1_data['baseDate'].dt.strftime("%H00")
PTY_data['baseTime'] = PTY_data['baseDate'].dt.strftime("%H00")
T1H_data['baseTime'] = T1H_data['baseDate'].dt.strftime("%H00")
REH_data['baseTime'] = REH_data['baseDate'].dt.strftime("%H00")
WSD_data['baseTime'] = WSD_data['baseDate'].dt.strftime("%H00")
#SKY_data['baseTime'] = SKY_data['baseDate'].dt.strftime("%H00")
```

```python
RN1_data['baseDate'] = RN1_data['baseDate'].dt.strftime("%Y%m%d")
PTY_data['baseDate'] = PTY_data['baseDate'].dt.strftime("%Y%m%d")
T1H_data['baseDate'] = T1H_data['baseDate'].dt.strftime("%Y%m%d")
REH_data['baseDate'] = REH_data['baseDate'].dt.strftime("%Y%m%d")
WSD_data['baseDate'] = WSD_data['baseDate'].dt.strftime("%Y%m%d")
#SKY_data['baseDate'] = SKY_data['baseDate'].dt.strftime("%Y%m%d")
```

```python
del RN1_data['day']
del PTY_data['day']
del T1H_data['day']
del REH_data['day']
del WSD_data['day']
```

```python
sample_data = pd.merge(left=RN1_data, right=PTY_data, how='left', on=['baseDate', 'baseTime'], sort=False)
sample_data = pd.merge(left=sample_data, right=T1H_data, how='left', on=['baseDate', 'baseTime'], sort=False)
sample_data = pd.merge(left=sample_data, right=REH_data, how='left', on=['baseDate', 'baseTime'], sort=False)
sample_data = pd.merge(left=sample_data, right=WSD_data, how='left', on=['baseDate', 'baseTime'], sort=False)
```

```python
sample_data = sample_data.reset_index(drop=True)
sample_data['nx'] = '58'
sample_data['ny'] = '125'
sample_data = sample_data[['baseDate', 'baseTime', 'T1H', 'RN1', 'REH', 'PTY', 'WSD', 'nx', 'ny']]
```

```python
sample_data.to_csv("C:/Users/happytoseeyou/Desktop/weather/58_125_weather.csv", encoding='utf-8-sig', index=False)
```
